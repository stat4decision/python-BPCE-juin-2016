```python
#importer donn√©es iris
iris=pd.read_excel("iris.xlsx")
```

```python
iris.head()
```

```python
#repr√©senter les deux premi√®res variables
plt.plot(iris[iris.columns[0]],iris[iris.columns[1]],"b.")
plt.title("Les iris")
plt.xlabel(iris.columns[0])
plt.ylabel(iris.columns[1])
```

```python
iris_group=iris.groupby("Espece")
colors = ['b', 'r', 'g']
```

```python
#repr√©sentation des esp√®ces
color_index=0
for group in iris_group:
    plt.plot(group[1][iris.columns[0]],group[1][iris.columns[2]],".",
             color=colors[color_index],label=group[0])
    color_index+=1
plt.legend()
```
